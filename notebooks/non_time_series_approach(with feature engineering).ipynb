# Overview

kaggle competition: https://www.kaggle.com/c/walmart-recruiting-store-sales-forecasting/overview/evaluation

In this recruiting competition, job-seekers are provided with historical sales data for 45 Walmart stores located in different regions. Each store contains many departments, and participants must project the sales for each department in each store. To add to the challenge, selected holiday markdown events are included in the dataset. These markdowns are known to affect sales, but it is challenging to predict which departments are affected and the extent of the impact.

## Create Helper Functions

 > First, two plot functions that will help us.

> The discrete plot is for finite numbers. We will use boxplot, to see the medians and interquartile ranges, and the striplot, which is a better way of seeing the distribution, even more when lots of outliers are present.

> The continuous plot, as the name says, is for continuous variables. We will see the distribution of probabilities and use BoxCox to understand if there is increase of correlation and decrease of skewness for each variable. In some cases the process of transforming a variable can help, depending on the model.

````python
def make_discrete_plot(dataframe, feature):
    fig = plt.figure(figsize=(20,8))
    gs = GridSpec(1,2)
    sns.boxplot(y=dataframe.Weekly_Sales, x=dataframe[feature], ax=fig.add_subplot(gs[0,0]))
    plt.ylabel('Sales', fontsize=16)
    plt.xlabel(feature, fontsize=16)
    sns.stripplot(y=dataframe.Weekly_Sales, x=dataframe[feature], ax=fig.add_subplot(gs[0,1]))
    plt.ylabel('Sales', fontsize=16)
    plt.xlabel(feature, fontsize=16)
````

````python
def make_continuous_plot(dataframe, feature):
    
    fig = plt.figure(figsize=(18,15))
    gs = GridSpec(2,2)
    
    j = sns.scatterplot(y=dataframe['Weekly_Sales'], 
                        x=boxcox1p(dataframe[feature], 0.15), ax=fig.add_subplot(gs[0,1]), palette = 'blue')

    plt.title('BoxCox 0.15\n' + 'Corr: ' + str(np.round(dataframe['Weekly_Sales'].corr(boxcox1p(dataframe[feature], 0.15)),2)) +
              ', Skew: ' + str(np.round(stats.skew(boxcox1p(dataframe[feature], 0.15), nan_policy='omit'),2)))
    
    j = sns.scatterplot(y=dataframe['Weekly_Sales'], 
                        x=boxcox1p(dataframe[feature], 0.25), ax=fig.add_subplot(gs[1,0]), palette = 'blue')

    plt.title('BoxCox 0.25\n' + 'Corr: ' + str(np.round(dataframe['Weekly_Sales'].corr(boxcox1p(dataframe[feature], 0.25)),2)) +
              ', Skew: ' + str(np.round(stats.skew(boxcox1p(dataframe[feature], 0.25), nan_policy='omit'),2)))
    
    j = sns.histplot(dataframe[feature], ax=fig.add_subplot(gs[1,1]), color = 'green')

    plt.title('Distribution\n')
    
    j = sns.scatterplot(y=dataframe['Weekly_Sales'], 
                        x=dataframe[feature], ax=fig.add_subplot(gs[0,0]), color = 'red')

    plt.title('Linear\n' + 'Corr: ' + str(np.round(dataframe['Weekly_Sales'].corr(dataframe[feature]),2)) + ', Skew: ' + 
               str(np.round(stats.skew(dataframe[feature], nan_policy='omit'),2)))
````

````python
def make_correlations_plot(dataframe):
    sns.set(style="white")

    corr = dataframe.corr()

    mask = np.triu(np.ones_like(corr, dtype=np.bool))

    f, ax = plt.subplots(figsize=(20, 15))

    cmap = sns.diverging_palette(220, 10, as_cmap=True)

    plt.title('Correlation Matrix', fontsize=18)

    sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
                square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)
````

On the other hand: TS data of some stores contain much more noise compared to others; some represents climbing growth trend, while others vice-versa; there are few stores with individual seasonal factor.

So why it is important to know the seasonal and trend components of our data set for different stores? 

- 1st scenario

    We could use the `SARIMAX` model, which is a handy and popular machine learning algorithm for TS predictions. This algorithm could be an alternative if we would make the overall sales predictions of Walmart, however, our purpose is to make forecasts for every store and every department. `SARIMAX` model implies only one target feature at a time. So, one way is to build a model and then loop it through each target feature. It could work in case if all target features would have identical seasonal and trend patterns. So with the power of the `pmdarima` library, we could approach the `auto_arima` function and specify p,d,q,P,D,Q parameters and loop the model all over the target features. However, in some stores, the picture is different from others. It means that we have to specify the individual parameters for particular stores and departments and use the `autoarima` function multiple times, which is not handy in terms of time and computational resources. Thus will drop this scenario for this case straight at a time.


- 2nd scenario

    `VARIMAX` algorithm is also an interesting tool in terms of predicting TS data with multiple target features at a time, which is our case. However, it shows the best performance, while targets have a particular level of causality. In our case, there are some stores that represent their own development of sales volume story. Thus we will skip this algorithm as well.

The two interesting models that is recommended to apply to our dataset are `Random Forest` and `RNN`.
Both can work in terms of prediction of multiple targets of sequential data and consider individuality for every target (in our case for every department of every store).

# Models

````python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import ExtraTreesRegressor, RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.feature_selection import RFE
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import KFold
````

Here we do simple (non time series) regression using several models.

````python
def knn():
    knn = KNeighborsRegressor(n_neighbors=10, n_jobs=-1)
    return knn

def extraTreesRegressor():
    clf = ExtraTreesRegressor(n_estimators=100,max_features='auto', verbose=1, n_jobs=-1)
    return clf

def randomForestRegressor():
    clf = RandomForestRegressor(n_estimators=100,max_features='log2', verbose=1, n_jobs=-1)
    return clf

def svm():
    clf = SVR(kernel='rbf', gamma='auto')
    return clf

def nn():
    clf = MLPRegressor(hidden_layer_sizes=(10,),  activation='relu', verbose=3)
    return clf

def predict_(m, test_x):
    return pd.Series(m.predict(test_x))

def model_(name:str='extraTreesRegressor'):
    if name == 'extraTreesRegressor':
        return extraTreesRegressor()
    elif name == 'knn':
        return knn()
    elif name == 'svm':
        return svm()
    elif name == 'nn':
        return nn()
    elif name == 'randomForestRegressor':
        return randomForestRegressor()
    
def train_(train_x, train_y, model_name):
    m = model_(model_name)
    m.fit(train_x, train_y)
    return m

def train_and_predict(train_x, train_y, test_x, model_name):
    m = train_(train_x, train_y, model_name)
    return predict_(m, test_x), m
````

We use mean absolute error for metric for now

````python
def calculate_error(test_y, predicted, weights):
    return mean_absolute_error(test_y, predicted, sample_weight=weights)
````

# K-Fold Cross Validation

We apply regular k-fold cross validation since our assumption is that data is static (non time series)

## Prepare Folds

````python
kf = KFold(n_splits=5)
splited = []

for name, group in train_ds.groupby(["Store", "Dept"]):
    group = group.reset_index(drop=True)
    trains_x = []
    trains_y = []
    tests_x = []
    tests_y = []
    if group.shape[0] <= 5:
        f = np.array(range(5))
        np.random.shuffle(f)
        group['fold'] = f[:group.shape[0]]
        continue
    fold = 0
    for train_index, test_index in kf.split(group):
        group.loc[test_index, 'fold'] = fold
        fold += 1
    splited.append(group)

splited = pd.concat(splited).reset_index(drop=True)
````

````python
splited
````

## Train Loop

````python
model_names = ['extraTreesRegressor','knn', 'nn','randomForestRegressor']
````

````python
best_models = {}
cv_errors = {}

for model_name in model_names:
    
    best_model = None
    error_cv = 0
    best_error = np.iinfo(np.int32).max
    
    for fold in range(5):
        
        print(f'Model: {model_name}\t Dataset Fold:{int(fold)}/5')
        
        dataset_train = splited.loc[splited['fold'] != fold]
        dataset_test = splited.loc[splited['fold'] == fold]
        
        train_y = dataset_train['Weekly_Sales']
        train_x = dataset_train.drop(columns=['Weekly_Sales', 'fold', 'Date'])
        test_y = dataset_test['Weekly_Sales']
        test_x = dataset_test.drop(columns=['Weekly_Sales', 'fold', 'Date'])
        
        print(f'Train: {dataset_train.shape}\t Test: {dataset_test.shape}')

        predicted, model = train_and_predict(train_x, train_y, test_x, model_name)
        
        # add weights for calculating error (for holidays we have higher weights!)
        weights = test_x['IsHoliday'].replace(True, 5).replace(False, 1)
        
        error = calculate_error(test_y, predicted, weights)
        error_cv += error

        print(f'Error: {error}')

        if error < best_error:
            print('Best model found!')
            best_error = error
            best_model = model
            
    error_cv /= 5
    
    best_models[model_name] = best_model
    cv_errors[model_name] = error_cv
````
